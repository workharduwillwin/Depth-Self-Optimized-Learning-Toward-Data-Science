```python
import numpy as np 
import matplotlib.pyplot as plt

policy_error_record=np.load("policy_error_record.npy")
td_error_record=np.load("td_error_record.npy")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.grid()
plt.title("Policy function")
plt.plot(policy_error_record,label="Policy error")
plt.xlabel("Episode")
plt.ylabel("Loss value")
plt.legend()
plt.subplot(1,2,2)
plt.grid()
plt.title("Value function")
plt.plot(td_error_record,label="TD error")
plt.xlabel("Episode")
plt.ylabel("Loss value")
plt.legend()
plt.savefig("p_loss_and_t_loss.jpg")
plt.show()
```
